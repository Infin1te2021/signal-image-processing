```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft, fftfreq, fftshift, ifftshift
```

fft -> Return discrete Fourier transform of real or complex sequence.
\begin{align*}
      & y[k] = \sum_{n=0}^{N-1}e^{-2\pi jk\frac{kn}{N}}x[n]&
\end{align*}
ifft -> Return discrete inverse Fourier transform of real or complex sequence.
\begin{align*}
      & x[n] = \frac{1}{N}\sum_{n=0}^{N-1}e^{2\pi jk\frac{kn}{N}}y[k]&
\end{align*}


```python
x = np.array([1.0, 2.0, 1.0, -1.0, 1.5])
y = fft(x)
y
```

```python
yinv = ifft(y)
yinv
```

```python
N = 600 # Number of sample points
T = 1.0 / 800.0 # sample spacing
x = np.linspace(start=0.0, stop=N*T, num=N) # Returns num evenly spaced samples, calculated over the interval [start, stop].
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x) 
yf = fft(y)
yif = ifft(yf)

plt.figure(figsize=(12, 8))

plt.subplot(3, 2, 1)
plt.plot(x, y)
plt.title('Original Signal')
plt.grid()

plt.subplot(3, 2, 3)
xf = np.linspace(0.0, 1.0 / (2.0 * T), N // 2)
plt.plot(xf, 2.0 / N * np.abs(yf[0:(N // 2)]))
plt.title('Frequency Domain')
plt.grid()

plt.subplot(3, 2, 5)
plt.plot(x, yif.real, label='Real')
plt.title('Inverse Fourier Transform (Real)')
plt.grid()

plt.subplot(3, 2, 6)
plt.plot(x, yif.imag, label='Imaginary')
plt.title('Inverse Fourier Transform (Imaginary)')
plt.grid()

plt.tight_layout()
plt.show()
```
