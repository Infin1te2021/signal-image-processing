```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

img = Image.open('src/chest-x-ray.png')
```

```python
plt.imshow(img, cmap='gray')
plt.show()
```

```python
# convert image into a numpy array
img = np.asarray(img)

# show the histogram
plt.hist(img, bins=256)
plt.show()
```

```python
img_flatten = img.flatten().reshape(1280, 1113)
plt.imshow(img_flatten, cmap='gray')
plt.show()
```

```python
plt.hist(img.flatten(), bins=256)
plt.show()
```
